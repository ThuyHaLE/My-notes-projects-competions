##String removal

```python
import re
```

```python
#Removing html and urls
text = ':-)) @82476 🤔🧐 🤓 😎 🥸 🤩 🥳 😏 😒 😞 😔 😟 😕 🙁 ☹️https://www.sciencedaily.com/releases/2023/08/230830131650.htmUniversity of Oxford researchers have made a significant step towards realising miniature bio-integrated devices, capable of directly stimulating cells. The work has been published today in the journal Nature.'
# import regex

def remove_html(text):
    html_pattern = re.compile('<.*?>')
    return html_pattern.sub(r' ', text)
def remove_urls(text):
    url_pattern = re.compile(r'https?://\S+|www\.\S+')
    return url_pattern.sub(r' ', text)

no_urls_string = remove_urls(remove_html(text))
no_urls_string
```

```python
#Removing all punctuation
import string
print([punc for punc in string.punctuation])

text = "The team attached the device to droplets containing human neural progenitor cells, which had been stained with a fluorescent dye to indicate their activity. When the power source was turned on, time-lapse recording demonstrated waves of intercellular calcium signalling* in the neurons, induced by the local ionic current. Dr Yujia Zhang (Department of Chemistry, University of Oxford), the lead researcher for the study, said: 'The miniaturized soft power source represents a breakthrough in bio-integrated devices. By harnessing ion gradients, we have developed a miniature, biocompatible system for regulating cells and tissues on the microscale, which opens up a wide range of potential applications in biology and medicine.' "
replace_punc_string = text
for punc in string.punctuation.replace("'",''):
  replace_punc_string = replace_punc_string.replace(punc, '')
replace_punc_string
```

```python
#Removing all punctuation except dot ([^\w\s.]), words and space
no_punc_string = re.sub(r"[^\w\s'.]", "", replace_punc_string)
no_punc_string
```

```python
#Finding and replace English Contractions
!pip -q install contractions
import contractions
text = '''I'll be there within 5 min. Shouldn't you be there too?
          I'd love to see u there my dear. It's awesome to meet new friends.
          We've been waiting for this day for so long.'''
no_contractions = contractions.fix(text)
no_contractions
```

```python
#Finding abbreviations in the text
dict_look = {'DM':'direct message', 'info':'information'}
def stand_text(text):
    for k, v in dict_look.items():
      text = text.replace(k, v)
    return text

text = "New DM: We've been waiting for this day for so long."
stand_text(text)
```

```python
#Removing white spaces
text = "New DM: We've been waiting    for this    day for so long."
no_wspace_string = re.sub(' + ', ' ', text).strip()
no_wspace_string
```

```python
#Removing numbers at the beginning and end of a string
text = "New DM: We've been waiting    for this    day for so long. :-)) @82476"
no_nums_string = re.sub(r'^[0-9]\d+\b|[0-9]\d+\b$', "",text).strip()
no_nums_string
```

# Basic Preprocessing (token based preprocessing)

```python
import nltk
#import nltk for tokenization
nltk.download('punkt')
```

```python
from nltk import word_tokenize

text = '''I'll be there within 5 min. Shouldn't you be there too?
          I'd love to see u there my dear. It's awesome to meet new friends.
          We've been waiting for this day for so long.'''

print(f'Token list: {[token.lower() for token in word_tokenize(text)]}')
```

```python
from nltk.tokenize import TreebankWordTokenizer

text = '''I'll be there within 5 min. Shouldn't you be there too?
          I'd love to see u there my dear. It's awesome to meet new friends.
          We've been waiting for this day for so long.'''

print(f'Token list: {[token.lower() for token in TreebankWordTokenizer().tokenize(text)]}')
```

```python
from nltk.tokenize import RegexpTokenizer

text = '''I'll be there within 5 min. Shouldn't you be there too?
          I'd love to see u there my dear. It's awesome to meet new friends.
          We've been waiting for this day for so long.'''

tokenizer = RegexpTokenizer(r'\w+|\$[\d.]+|.\S+') #\$\s*\d+

print(f'Token list: {[w.lower() for w in tokenizer.tokenize(text)]}')
```

##Stop Word Removal

```python
#download stopwords
nltk.download('stopwords')
```

```python
#Removing Stop Words
#import nltk for stopwords
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))

text = '''I'll be there within 5 min. Shouldn't you be there too?
          I'd love to see u there my dear. It's awesome to meet new friends.
          We've been waiting for this day for so long.'''

print(f'Stop_words: {[w for w in stop_words]}')

#remove stopwords
print([i for i in [w.lower() for w in word_tokenize(text)] if i in stop_words])
no_stpwords_string=" ".join([i for i in [w.lower() for w in word_tokenize(text)] if not i in stop_words])
print(no_stpwords_string)

#removing last space
no_stpwords_string = no_stpwords_string[:-1]
no_stpwords_string
print(no_stpwords_string)
```

##Lemmatization

```python
#Lemmatization
from nltk.tokenize import TreebankWordTokenizer
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
text = "He told Dr. Lovato that he was done with the tests and would post the results shortly."

#lemmatizer
print(f'Token list: {[token for token in TreebankWordTokenizer().tokenize(text)]}')
print(f'Apply lemmatization: {[f"({t} => {lemmatizer.lemmatize(t)})" for t in [token.lower() for token in TreebankWordTokenizer().tokenize(text)]]}')
```

##Stemmer

```python
#Stemming
from nltk.tokenize import TreebankWordTokenizer
from nltk.stem.porter import PorterStemmer
stemmer = PorterStemmer()

text = "He told Dr. Lovato that he was done with the tests and would post the results shortly."

#Stemming
print(f'Token list: {[token for token in TreebankWordTokenizer().tokenize(text)]}')
print(f'Apply lemmatization: {[f"({t} => {stemmer.stem(t)})" for t in [token.lower() for token in TreebankWordTokenizer().tokenize(text)]]}')
```

##Rate words

##Part-of-Speech Tagging

```python
#Part-of-Speech Tagging
text = '''I'll be there within 5 min. Shouldn't you be there too?
          I'd love to see u there my dear. It's awesome to meet new friends.
          We've been waiting for this day for so long.'''

pos_tagged_text = nltk.pos_tag(text.split())
print(f'Apply pos_tagged: {[f"({w} => {t})" for w, t in pos_tagged_text]}')
```

```python
#Lemmatizer
nltk.download('wordnet') #Resource for wordnet
nltk.download('averaged_perceptron_tagger') #Resource for nltk.pos_tag()
```

```python
#Combine Lemmatizer with Part-of-Speech Tagging
from nltk.tokenize import TreebankWordTokenizer
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet

text = "He told Dr. Lovato that he was done with the tests and would post the results shortly."

lemmatizer = WordNetLemmatizer()
wordnet_map = {'N':wordnet.NOUN,'J':wordnet.ADJ,'V':wordnet.VERB,'R':wordnet.ADV}
print(wordnet_map)

pos_tagged_text = nltk.pos_tag(text.split())
print(f'Apply pos_tagged: {[f"({w} => {t})" for w, t in pos_tagged_text]}')
print(f'Apply pos_tagged: {[f"({w} => {t[0]} => {wordnet_map.get(t[0], wordnet.VERB)})" for w, t in pos_tagged_text]}')

#lemmatize without condition
print([f"({word} => {lemmatizer.lemmatize(word)})" for word in TreebankWordTokenizer().tokenize(text)])

#lemmatize with condition (a denotes verb in "pos")
print([f"({word} => {lemmatizer.lemmatize(word, wordnet_map.get(pos[0], wordnet.VERB))}" for word, pos in pos_tagged_text])
```
